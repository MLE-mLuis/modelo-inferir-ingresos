##5- Detección y eliminación de valores atípicos

Gráficos BOXPLOT

```python
#Creación de función para realizar gráfico de BOXPLOT
def Grafico_BOXPLOT(df_in):
  df = df_in.copy()

  plt.figure(figsize=(15,10))
  gs = gridspec.GridSpec(3, 2)
  gs.update(hspace=0.8)

  for i, f in enumerate(df):
      q1 = df[f].quantile(0.25)
      q3 = df[f].quantile(0.75)
      iqr = q3-q1 #Interquartile range
      rango_inicio = q1-1.5*iqr
      rango_inicio_str = '{0:.2f}'.format(rango_inicio)
      rango_fin = q3+1.5*iqr
      rango_fin_str = '{0:.2f}'.format(rango_fin)
      
      ax = plt.subplot(gs[i])
      sns.boxplot(df[f])
      ax.set_xlabel('')
      ax.set_title('feature: ' + str(f) + ' | IQR [ '+ str(rango_inicio_str) +' - '+ str(rango_fin_str)  +' ]')

  plt.show()
```

```python
#Creación de función para eliminar valores atípicos (considerando el IQR)
def Remove_Outlier(df_in, col_name):
  q1 = df_in[col_name].quantile(0.25)
  q3 = df_in[col_name].quantile(0.75)
  iqr = q3-q1 #Interquartile range
  rango_inicio = q1-1.5*iqr
  rango_fin = q3+1.5*iqr
  print('Columna: ',col_name)
  print('  IQR - Rango Inicio: ',rango_inicio)
  print('  IQR - Rango Fin: ',rango_fin)
  df_out = df_in.loc[(df_in[col_name] > rango_inicio) & (df_in[col_name] < rango_fin)]

  registros_eliminados = df_in.shape[0] - df_out.shape[0]
  
  print('Registros')
  print('  Inicial:    ', '{:,}'.format(df_in.shape[0]) ,"-")
  print('  Eliminados:   ','{:,}'.format(registros_eliminados))
  print('  ----------------------')
  print('  Final:      ', '{:,}'.format(df_out.shape[0]))

  return df_out
```

**Pasos importantes:**
* Describir el modelo ---> glm()
* Ajuste del modelo ---> .fit()
* Resumen del modelo ---> .summary()
* Calculo de predicciones --- > .predict()

**Forma general de los modelos:**
* model = glm(formula, data, family)

**Las formulas pueden contener:**
* C(x1), solo si x1 es una variable categorica
* -1, solo si el modelo no necesita el intercepto
* x1:x2, es la parte de interacción entre las variables x1 y x2
* x1*x2, ademas de la interacción añade el efecto individual de x1 y x2
* np.log(x1), solo si es necesario una transformación en las variables.

**Las familias son: family = sm.families.___()**
* Gaussian(link = sm.families.links.identity) ---> por defecto
* Binomial(link = sm.families.links.logit) ---> logistica
      - probit, cauchy, log, cloglog
* Poisson(link = sm.families.links.log)
      - identity, sqrt
* Entre otros

**Pasos importantes:**
* Describir el modelo ---> glm()
* Ajuste del modelo ---> .fit()
* Resumen del modelo ---> .summary()
* Calculo de predicciones --- > .predict()

**Forma general de los modelos:**
* model = glm(formula, data, family)

**Las formulas pueden contener:**
* C(x1), solo si x1 es una variable categorica
* -1, solo si el modelo no necesita el intercepto
* x1:x2, es la parte de interacción entre las variables x1 y x2
* x1*x2, ademas de la interacción añade el efecto individual de x1 y x2
* np.log(x1), solo si es necesario una transformación en las variables.

**Las familias son: family = sm.families.___()**
* Gaussian(link = sm.families.links.identity) ---> por defecto
* Binomial(link = sm.families.links.logit) ---> logistica
      - probit, cauchy, log, cloglog
* Poisson(link = sm.families.links.log)
      - identity, sqrt
* Entre otros

**Forma general de los modelos:**
* model = glm(formula, data, family)

**Las formulas pueden contener:**
* C(x1), solo si x1 es una variable categorica
* -1, solo si el modelo no necesita el intercepto
* x1:x2, es la parte de interacción entre las variables x1 y x2
* x1*x2, ademas de la interacción añade el efecto individual de x1 y x2
* np.log(x1), solo si es necesario una transformación en las variables.

**Las familias son: family = sm.families.___()**
* Gaussian(link = sm.families.links.identity) ---> por defecto
* Binomial(link = sm.families.links.logit) ---> logistica
      - probit, cauchy, log, cloglog
* Poisson(link = sm.families.links.log)
      - identity, sqrt
* Entre otros
